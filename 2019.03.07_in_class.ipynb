## Random Generator: Uniform probability

* random([size]):	Return random floats in the half-open interval [0.0, 1.0).

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.random(1000)
#print(x)
```

## Plot Histogram

```python
plt.hist(x,10)

plt.xlabel("x value")
plt.ylabel("Numbers of elements")
plt.show()
```

## Random Generator: Gaussian probability

* normal([loc, scale, size]):	Draw random samples from a normal (Gaussian) distribution.

```python
mu, sigma = 0, 0.1 # mean and standard deviation
s = np.random.normal(mu, sigma, 1000)
#print(s)
plt.hist(s,10)

#plt.xlim(-2,2)
plt.show()
```

```python
ss=plt.hist(s,10)
print(ss)
```

* Create regular bins for the histogram using np.linspace

```python
bins=np.linspace(-0.5, 0.5, 21, endpoint=True)
print(bins)
plt.hist(s,bins)

#plt.xlim(-2,2)
plt.show()
```

## Add noise to Blackbody curve

```python
import numpy as np
import matplotlib.pyplot as plt
import math

#SI unit
h = 6.62*10**-34 #Planck constant
c = 3*10**8      #speed of light
k = 1.38*10**-23 #Boltzmann Constant JK^-1

num = 100      # total number of points

v = np.linspace(10**0, 10**15, num, endpoint=True)

#T=float(input("Input temperature:"))
T=5500.
B=2*h*v**3/c**2/(math.e**(h*v/k/T)-1)

#plt.xlim(10**11, 10**15.2)
#plt.ylim(10**-14,10**-6)
plt.plot(v,B,".")
#plt.loglog(v,B,".")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Intensity Watts/Hz/m^2")

dB=np.random.normal(0., 10**-8.5,num)
B += dB

plt.plot(v,B)
#plt.loglog(v,B)

plt.show()                
```
