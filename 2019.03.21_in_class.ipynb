```python
%matplotlib inline
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np

im1 = mpimg.imread('images/IMG_3827.JPG')
plt.imshow(im1)
```

```python
im2 = mpimg.imread('images/IMG_3828.JPG')
plt.imshow(im2)
im3 = mpimg.imread('images/IMG_3829.JPG')
plt.imshow(im3)
im4 = mpimg.imread('images/IMG_3830.JPG')
plt.imshow(im4)
im5 = mpimg.imread('images/IMG_3831.JPG')
plt.imshow(im5)
```

#### Read the blue image of the first image

```python
b_im1 = im1[:, :, 2]
print(b_im1)
print(b_im1.shape)
```

#### Stack two blue images together to form a new np array

```python
b_im2 = im2[:, :, 2]
b_im = np.stack((b_im1, b_im2),axis=0)
print(b_im.shape)
print(b_im)
```

#### Take average of the stacked array in the 0th direction

```python
# Plot average of two images
print(np.mean(b_im))               #   average all element => one number 
b_ave = np.mean(b_im, axis=0)      #   axis = 0 indicate the direction for averaging => one array
plt.imshow(b_ave)
```

#### Take median value of the stacked array in the 0th direction

```python
b_median = np.median(b_im, axis=0)
plt.imshow(b_median)
```

# Now plot the average image and the median image of your 5 images

```python
b_im = np.stack((im1[:, :, 2], im2[:, :, 2],im3[:, :, 2],im4[:, :, 2],im5[:, :, 2]))
b_ave = np.mean(b_im, axis=0)
plt.imshow(b_ave)
```

```python
im = np.stack((im1, im2,im3,im4,im5),axis=0)
print(im.shape)
im_ave = np.mean(im, axis=0)
print(im_ave.shape)
#print(im_ave.astype(int))
#plt.imshow(im_ave)
print(type(im_ave))
plt.imshow(im_ave.astype(np.uint8))
#plt.imshow(im_ave.astype(int))
plt.axis("off")
plt.savefig("one_5th_of_Travis.JPG",bbox_inches='tight', pad_inches=0)
```
